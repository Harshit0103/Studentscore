```python
#import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv('StudentsPerformance.csv')  #dataframe calling CSV file 
print (df.head())     #top5 entries
```

# Gender distribution

```python
plt.figure(figsize=(4,4))  #Shape the size 
ax = sns.countplot(data = df, x = "gender")
ax.bar_label(ax.containers[0])  #Show the total size
plt.title("Gender distribution")
plt.show()
```

# From the above chart we analyze that the numbers of females are bigger than males

```python
#now analyzing the impact of parent level education on all scores 

gb = df.groupby('parental level of education').agg({'math score': "mean", 'reading score': "mean", 'writing score': "mean"})
print(gb)
```

```python
sns.heatmap(gb)    # with heatmap we can define the values 
plt.show() 
```

```python
sns.heatmap(gb, annot = True)   # showing the values 
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb, annot = True) 
plt.title("parent Education Level")
plt.show()
```

# By the above chart it clears that parental education's have good impact on students score

```python
#analyzing the impact of race/ethnicity on the scores 
ht = df.groupby('race/ethnicity').agg({'math score': "mean", 'reading score': "mean", 'writing score': "mean"})
print(ht)
```

```python
sns.heatmap(ht)    
plt.show() 
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(ht)    
plt.show() 
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(ht, annot= True)    
plt.title("ethnicity distribution")
plt.show() 
```

# In the above chart we analyze that group E has very big difference in compare to other groups in the students score
